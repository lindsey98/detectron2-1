```python
from pycocotools.coco import COCO
from pycocotools.cocoeval import COCOeval
import pickle
```

```python
import os
os.chdir('..')
```

```python
def print_res(itera):
    with open('./output/adv_%s/logo_testacc.pkl'%str(itera), 'rb') as f:
        logo_testacc = pickle.load(f)
    
    with open('./output/adv_%s/input_testacc.pkl'%str(itera), 'rb') as f:
        input_testacc = pickle.load(f)

    with open('./output/adv_%s/all_testacc.pkl'%str(itera), 'rb') as f:
        all_testacc = pickle.load(f)
        
    print('-'*20, 'Clean testing results', '-'*20)
    print('*'*15, 'All classes', '*'*15)
    all_testacc.evaluate()
    all_testacc.accumulate()
    all_testacc.summarize()
    
    print('*'*15, 'Input class', '*'*15)
    input_testacc.evaluate()
    input_testacc.accumulate()
    input_testacc.summarize()

    print('*'*15, 'Logo class', '*'*15)
    logo_testacc.evaluate()
    logo_testacc.accumulate()
    logo_testacc.summarize()
    
    with open('./output/adv_%s/logo_perturb_testacc.pkl'%str(itera), 'rb') as f:
        logo_testacc = pickle.load(f)

    with open('./output/adv_%s/input_perturb_testacc.pkl'%str(itera), 'rb') as f:
        input_testacc = pickle.load(f)

    with open('./output/adv_%s/all_perturb_testacc.pkl'%str(itera), 'rb') as f:
        all_testacc = pickle.load(f)
        
    print('-'*20, 'Perturbed testing results', '-'*20)
    print('*'*15, 'All classes', '*'*15)
    all_testacc.evaluate()
    all_testacc.accumulate()
    all_testacc.summarize()
    
    print('*'*15, 'Input class', '*'*15)
    input_testacc.evaluate()
    input_testacc.accumulate()
    input_testacc.summarize()

    print('*'*15, 'Logo class', '*'*15)
    logo_testacc.evaluate()
    logo_testacc.accumulate()
    logo_testacc.summarize()
    
    
    with open('./output/adv_%s/logo_trainacc.pkl'%str(itera), 'rb') as f:
        logo_trainacc = pickle.load(f)

    with open('./output/adv_%s/input_trainacc.pkl'%str(itera), 'rb') as f:
        input_trainacc = pickle.load(f)

    with open('./output/adv_%s/all_trainacc.pkl'%str(itera), 'rb') as f:
        all_trainacc = pickle.load(f)
        
    print('-'*20, 'Training results(D0+Di)', '-'*20)
    
    print('*'*15, 'All classes', '*'*15)
    all_trainacc.evaluate()
    all_trainacc.accumulate()
    all_trainacc.summarize()
    
    print('*'*15, 'Input class', '*'*15)
    input_trainacc.evaluate()
    input_trainacc.accumulate()
    input_trainacc.summarize()
    
    print('*'*15, 'Logo class', '*'*15)
    logo_trainacc.evaluate()
    logo_trainacc.accumulate()
    logo_trainacc.summarize()
```

```python
print_res(4)
```
