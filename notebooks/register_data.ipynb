```python
from detectron2.data.datasets import register_coco_instances
from pathlib import Path
```

```python
# Define dataset paths
data_dir = Path("data")

benign_data_dir = data_dir / "benign_data"
benign_img_dir = benign_data_dir / "benign_database"

benign_gt_coco_path = benign_data_dir / "coco_perturb_gt.json"

# Register benign train and test sets
register_coco_instances("benign_adv", {}, benign_gt_coco_path, benign_img_dir)
```

```python
from detectron2.data import MetadataCatalog
MetadataCatalog.get("benign_adv")
```

```python
from detectron2.config import get_cfg
from detectron2_1.datasets import BenignMapper
from detectron2.modeling import build_model
from detectron2.checkpoint import DetectionCheckpointer

cfg_path = 'configs/faster_rcnn_adv.yaml'
weights_path = 'rcnn_bet365.pth'
cfg_ = get_cfg()
cfg_.merge_from_file(cfg_path)
cfg_.MODEL.WEIGHTS = weights_path

# Modify config
cfg = cfg_.clone()  # cfg can be modified by model
# To generate more dense proposals

# Init model
model = build_model(cfg)
model.eval()

# Load weights
checkpointer = DetectionCheckpointer(model)
checkpointer.load(cfg.MODEL.WEIGHTS)
```

```python
from detectron2.data import DatasetMapper, MetadataCatalog, build_detection_test_loader
# Init dataloader on test dataset
mapper = BenignMapper(cfg, is_train=False)
data_loader = build_detection_test_loader(
    cfg, 'benign_adv', mapper=mapper
)
```

```python
for i, batch in enumerate(data_loader):
    print(batch)
    if i == 1:
        break
```
